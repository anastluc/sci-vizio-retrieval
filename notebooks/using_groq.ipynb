```python
# Function to make API request
def make_api_request(model):
    response = requests.post(
        GROQ_API_URL,
        json={
            "model": model,
            "messages": messages,
            "max_tokens": 1000
        },
        headers={
            "Authorization": f"Bearer {GROQ_API_KEY}",
            "Content-Type": "application/json"
        },
        timeout=30
    )
    return response
```
